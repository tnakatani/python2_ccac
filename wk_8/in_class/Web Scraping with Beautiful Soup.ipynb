```python
import urllib
from bs4 import BeautifulSoup

def getMomaURL(term):
    url = (f'https://www.moma.org/collection/works?locale=en&utf8=%E2%9C%93' +
            '&q={term}&classifications=any' +
            '&date_begin=Pre-1850&date_end=2020' +
            '&with_images=1')
    return url
    
def getMixcloudURL(term):
    url = f'https://www.mixcloud.com/{term}/'
    return url

def getGoodReadsURL(term):
    url = f'https://www.goodreads.com/search?query={term}'
    return url

def getPageText(url):
    req = urllib.request.Request(url)
    with urllib.request.urlopen(req) as response:
        return response.read()
```

```python
moma = getPageText(getMomaURL('Corbusier'))
moma
```

```python
moma_soup = BeautifulSoup(moma, 'html.parser')
caption = moma_soup.find_all('div','work--in-list__caption-container')
caption
```

```python
pageText = getPageText(getGoodReadsURL('python'))
pageText
```

```python
soup = BeautifulSoup(pageText, 'html.parser')
s = soup.find_all('a','bookTitle')
print(s)
totalTitles = 0
subTitles = 0

for item in s:
    t = item.find('span').string
    print(t)
    totalTitles = totalTitles + 1
    if ":" in t:
        subTitles = subTitles + 1
print("********SUMMARY**********")
print("Total Titles: ", str(totalTitles))
print(str((subTitles/totalTitles)*100),"% of Titles have subtitles")
```
